```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

diabetes = load_diabetes()
b_pressure = diabetes.data[:, 2]
b_glucose = diabetes.data[:, 3]
b_glucose = b_glucose .reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(b_glucose, b_pressure, test_size=0.33, random_state=42)

model = LinearRegression().fit(X_train, y_train)

predict = model.predict(X_test)

print('(A) Intercept: %.3f' % model.intercept_)
print('(B) Coefficient: %.3f' % model.coef_[0])
print("(G) R² Score: %.3f" % r2_score(y_test, predict))
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, model.coef_[0] * X_train + model.intercept_, '-r')
plt.xlabel("Blood Pressure")
plt.ylabel("Blood Glucose")
plt.show()
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, model.coef_[0] * X_test + model.intercept_, '-r')
plt.xlabel("Blood Pressure")
plt.ylabel("Blood Glucose")
plt.show()
```

```python
print("(D) Sum of Squared Errors (SSE): %.3f" % np.sum((y_test - predict) ** 2))
print("(E) Mean Squared Error (MSE): %.3f" % mean_squared_error(y_test, predict))
print("(F) Mean Absolute Error (MAE): %.3f" % mean_absolute_error(y_test, predict))
```
